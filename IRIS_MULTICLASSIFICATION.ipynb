```python
from sklearn.model_selection import train_test_split
train_x,test_x,train_y,test_y=train_test_split(x,y,test_size=0.2)
from sklearn.svm import SVC
model=SVC()
model.fit(train_x,train_y)
model.score(test_x,test_y)
```
